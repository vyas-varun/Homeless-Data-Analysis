```python
df = pd.ExcelFile('/content/2007_2020_PIT_State.xlsx')
df0 = pd.read_excel(df, '2020')
df1 = pd.read_excel(df, '2019')
df2 = pd.read_excel(df, '2018')
df3 = pd.read_excel(df, '2017')
df4 = pd.read_excel(df, '2016')
df5 = pd.read_excel(df, '2015')
df6 = pd.read_excel(df, '2014')
df7 = pd.read_excel(df, '2013')
df8 = pd.read_excel(df, '2012')
df9 = pd.read_excel(df, '2011')
df10 = pd.read_excel(df, '2010')
df11 = pd.read_excel(df, '2009')
df12 = pd.read_excel(df, '2008')
df13 = pd.read_excel(df, '2007')
```

**Complete Data Set of Homeless Population of States in 2020-2027**

```python
from functools import reduce
Complete_DataSet = [df0, df1, df2, df3, df4, df5,df6, df7, df8,df9, df10, df11,df12, df13]
Complete_DataSet = reduce(lambda  left,right: pd.merge(left,right,on=['State'],
                                            how='outer'), Complete_DataSet)
Complete_DataSet.head()
```

```python
#merging overall Homeless of 2020 - 2007
Overall_Homeless_2007_2020 = df0[["State", 
         "Overall Homeless, 2020"]].merge(df1[["State","Overall Homeless, 2019"]]).merge(df2[["State","Overall Homeless, 2018"]]).merge(df3[["State","Overall Homeless, 2017"]]).merge(df4[["State","Overall Homeless, 2016"]]).merge(df5[["State","Overall Homeless, 2015"]]).merge(df6[["State","Overall Homeless, 2014"]]).merge(df7[["State","Overall Homeless, 2013"]]).merge(df8[["State","Overall Homeless, 2012"]]).merge(df9[["State","Overall Homeless, 2011"]]).merge(df10[["State","Overall Homeless, 2010"]]).merge(df11[["State","Overall Homeless, 2009"]]).merge(df12[["State","Overall Homeless, 2008"]]).merge(df13[["State","Overall Homeless, 2007"]],
                                     on = "State", 
                                     how = "left")
  
```

```python
Overall_Homeless_2007_2020.head()
```

```python
#replacing blank values with NA
Overall_Homeless_2007_2020.replace(r'^\s*$', np.nan, regex=True)
```

```python
#converting object type to float 
cols = Overall_Homeless_2007_2020.columns
Overall_Homeless_2007_2020[cols[1:]] = Overall_Homeless_2007_2020[cols[1:]].apply(pd.to_numeric, errors='coerce')
Overall_Homeless_2007_2020.dtypes
```

```python
#creating column for Total overall Homeless from 2007 to 2020 
Overall_Homeless_2007_2020['Total_Homeless'] = Overall_Homeless_2007_2020.sum(axis=1,numeric_only= True)
Overall_Homeless_2007_2020.head()
```
